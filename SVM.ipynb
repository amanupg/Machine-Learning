```python
from sklearn.datasets import make_blobs
import matplotlib.pyplot as plt
```

```python
x, y = make_blobs(n_samples = 50, centers = 2, n_features=2, random_state=0, cluster_std=0.5)
plt.scatter(x[:,0],x[:,1], c=y)
plt.show()
```

```python
from sklearn.svm import SVC
model = SVC(kernel = 'poly', degree =8) #rbf - Radial Basis Function, poly- Polynomial, linear- Linear
model.fit(x,y)
```

```python
import numpy as np
xin = np.arange(-1,4)
yin = np.arange(-1,7)
Yin, Xin = np.meshgrid(yin, xin) #Meshgrid gives us the transpose
xy = np.vstack([Xin.ravel(), Yin.ravel()]).T
p = model.decision_function(xy).reshape(Xin.shape)
```

```python
plt.contour(Xin, Yin, p, levels=[-1,0,1])
plt.scatter(x[:,0], x[:,1], c=y)
plt.show()
```
