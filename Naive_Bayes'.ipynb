```python
import pandas as pd
df = pd.read_csv('/content/PlayTennis.csv')
df.head()
```

```python
total = df['Play Tennis'].value_counts()
df.groupby(['Outlook','Play Tennis']).size()
poy = 9/14
pon = 5/14
```

```python
outlook = pd.crosstab(df['Outlook'],df['Play Tennis'])
outlook['ProbOfYes'] = outlook['Yes']/outlook['Yes'].sum()
outlook['ProbOfNo'] = outlook['No']/outlook['No'].sum()
outlook
```

```python
temp = pd.crosstab(df['Temperature'],df['Play Tennis'])
temp['ProbOfYes'] = temp['Yes']/temp['Yes'].sum()
temp['ProbOfNo'] = temp['No']/temp['No'].sum()
temp
```

```python
humid = pd.crosstab(df['Humidity'],df['Play Tennis'])
humid['ProbOfYes'] = humid['Yes']/humid['Yes'].sum()
humid['ProbOfNo'] = humid['No']/humid['No'].sum()
humid
```

```python
wind = pd.crosstab(df['Wind'],df['Play Tennis'])
wind['ProbOfYes'] = wind['Yes']/wind['Yes'].sum()
wind['ProbOfNo'] = wind['No']/wind['No'].sum()
wind
```

```python
#P(Yes |Sunny, Hot, High, Weak) = P(y)*P(sunny y)*p(Hot y)*p(high y)*p(weak y)
#Therefore, probability P is:
pyes =poy* outlook.loc['Sunny','ProbOfYes'] * temp.loc['Hot','ProbOfYes'] * humid.loc['High','ProbOfYes'] * wind.loc['Weak','ProbOfYes']
pyes
```

```python
x1 = 'ProbOfNo'
pno = pon* outlook.loc['Sunny',x1] * temp.loc['Hot',x1] * humid.loc['High',x1] * wind.loc['Weak',x1]
pno
```

```python
#Normalised probability:
yon = (pyes)/(pyes+pno)
non = (pno)/(pyes+pno)
print(yon,non)
```

```python
#Using Scikit-:learn library:
x = df.iloc[:,:-1].values
y = df.iloc[:,-1].values
```

```python
#Converting the input into numerical values using encoding

from sklearn.preprocessing import OrdinalEncoder
enc = OrdinalEncoder()
x_enc = enc.fit_transform(x)
x_enc[0]
```

```python
from sklearn.naive_bayes import MultinomialNB
clf = MultinomialNB(alpha=0, fit_prior=False)
clf.fit(x_enc,y)
```

```python
pr = df.iloc[0,:-1].values
pr = enc.transform([pr])
pr
```

```python
print(clf.predict(pr))
clf.predict_proba(pr)
```
